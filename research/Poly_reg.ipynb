```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
```

```python
os.getcwd()
```

```python
df = pd.read_csv('../artifacts/data_transformation/train.csv')
df.head()
```

```python
df.shape
```

```python
X_train = df.drop(columns=['rent_amount'],axis=1)
y_train = df[['rent_amount']]
```

```python
X_train
```

```python
y_train
```

```python
poly = PolynomialFeatures(degree=2)
poly_feat = poly.fit_transform(X_train)
lin_reg = LinearRegression()
lin_reg.fit(poly_feat,y_train)
```

```python
test = pd.read_csv('../artifacts/data_transformation/test.csv')
test.head()
```

```python
test.shape
```

```python
X_test = test.drop(columns=['rent_amount'],axis=1)
y_test = test[['rent_amount']]
```

```python
X_test
```

```python
y_test
```

```python
pred = lin_reg.predict(poly.transform(X_test))
```

```python
r2_score(y_test,pred)
```

```python
r2_score(y_train, lin_reg.predict(poly.transform(X_train)))
```

```python
r2_score(y_test, lin_reg.predict(poly.transform(X_test)))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lin_reg, X_train, y_train, cv=5, scoring='r2')
print(scores)
print("Mean CV Score :", scores.mean())
```
